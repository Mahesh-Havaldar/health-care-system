```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
import joblib
```

```python
train_data = pd.read_csv("Training.csv")
test_data = pd.read_csv("Testing.csv")
```

```python
train_data.head()
```

```python
print("âœ… Training data shape:", train_data.shape)
print("âœ… Testing data shape:", test_data.shape)
```

```python
if "Unnamed: 133" in train_data.columns:
    train_data.drop("Unnamed: 133", axis=1, inplace=True)
```

```python
X = train_data.drop("prognosis", axis=1)
y = train_data["prognosis"]
```

```python
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

```python
X_train, X_val, y_train, y_val = train_test_split( X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)
```

```python
rf_model = RandomForestClassifier(n_estimators=200, random_state=42)
rf_model.fit(X_train, y_train)

# Decision Tree (for comparison)
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)
```

```python
rf_preds = rf_model.predict(X_val)
print("\nðŸ”¹ Random Forest Performance:")
print("Accuracy:", accuracy_score(y_val, rf_preds))
```

```python
dt_preds = dt_model.predict(X_val)
print("\nðŸ”¹ Decision Tree Performance:")
print("Accuracy:", accuracy_score(y_val, dt_preds))
```

```python
joblib.dump(rf_model, "disease_prediction_model.pkl")
joblib.dump(label_encoder, "label_encoder.pkl")
print("\nðŸŽ‰ Model and label encoder saved successfully!")
```
